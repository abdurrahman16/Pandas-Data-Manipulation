```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x= np.array([4,5,7,9,3,15])
y=np.array([6,3,8,15,14,16])
```

```python
plt.title("Test Plot View")
plt.xlabel("X Axis Data")
plt.ylabel("Y Axis Data")
plt.plot(x,y)
```

```python
plt.bar(x,y)
plt.legend(loc='upper right', title='Tester')
```
